```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
movies = pd.read_csv('python data/movies.csv')
ratings = pd.read_csv('python data/ratings.csv')
```

```python
data = pd.merge(ratings, movies, on='movieId')
```

```python
data.head()
```

```python
top_movies = data.groupby('title')['rating'].mean().sort_values(ascending=False).head(10)
top_movies
```
